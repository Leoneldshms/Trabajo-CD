# Eliminacion o impiutacion de valores faltates

```python
df1=df
```

```python
df1
```

```python
df1['Name'].fillna('No tiene', inplace=True)
df1['Name'].isnull().sum()
```

```python
df1['Name'].value_counts()
```

```python
df1['Age'].fillna('Sin edad', inplace=True)
df1['Age'] = pd.to_numeric(df1['Age'], errors='coerce')
df1['Age'].fillna(df1['Age'].mean(), inplace=True)
df1['Age'].isnull().sum()
```

```python
df1['Age'].value_counts()
```

```python
df1['Gender'].fillna('Otro', inplace=True)
df1['Gender'].isnull().sum()
```

```python
df1['Blood Type'].fillna('No tiene', inplace=True)
df1['Blood Type'].isnull().sum()
```

```python
df1['Medical Condition'].fillna('No especificada', inplace=True)
df1['Medical Condition'].isnull().sum()
```

```python
df1=df1.dropna(subset=['Date of Admission'])
df1['Date of Admission'].isnull().sum()
```

```python
df1['Doctor'].fillna('No tiene', inplace=True)
df1['Doctor'].isnull().sum()
```

```python
df1['Hospital'].fillna('No tiene', inplace=True)
df1['Hospital'].isnull().sum()
```

```python
df1['Insurance Provider'].fillna('No tiene', inplace=True)
df1['Insurance Provider'].isnull().sum()
```

```python
df1['Billing Amount'] = pd.to_numeric(df1['Billing Amount'], errors='coerce')
df1['Billing Amount'].fillna(df1['Billing Amount'].mean(), inplace=True)
df1['Billing Amount'].isnull().sum()
```

```python
df1['Room Number'] = pd.to_numeric(df1['Room Number'], errors='coerce')
df1['Room Number'].fillna(df1['Room Number'].mean(), inplace=True)
df1['Room Number'].isnull().sum()
```

```python
df1['Admission Type'].fillna('Otro', inplace=True)
df1['Admission Type'].isnull().sum()
```

```python
df1=df1.dropna(subset=['Discharge Date'])
df1['Discharge Date'].isnull().sum()
```

```python
df1['Medication'].fillna('No especificado', inplace=True)
df1['Medication'].isnull().sum()
```

```python
df1['Test Results'].fillna('En progreso', inplace=True)
df1['Test Results'].isnull().sum()
```
